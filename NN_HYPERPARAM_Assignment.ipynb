```
import numpy as np
import pandas as pd
import random
%matplotlib inline
import matplotlib.pyplot as plt
%tensorflow_version 2.x
import tensorflow as tf
from tensorflow import keras
import tensorflow.keras.backend as K
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.optimizers import SGD
```

1) number of epochs, number of neurons, batch size, optimizers

```
#2)

neuron_num = np.random.randint(40,250,50)
bsize = np.random.randint(5,25,50)
opt = ['Adadelta', 'Adagrad', 'Adam', 'Adamax', 'Ftrl', 
       'Nadam', 'RMSprop', 'SGD']
epoch_num = np.random.randint(9,15,50)



```

```
#3) and #4)
fash_mnist = keras.datasets.fashion_mnist
data = fash_mnist.load_data()
```

```
X_train, y_train, X_test, y_test = data[0][0], data[0][1], data[1][0], data[1][1]
X_train = X_train/255
X_test = X_test/255

X_train, y_train, X_validation, y_validation = X_train[:-15000], y_train[:-15000], X_train[-15000:], y_train[-15000:]
```

```
for _ in range(10):
  K.clear_session()
  
  ins = keras.Input(shape=(28,28,))
  a = keras.layers.Flatten()(ins)
  b = keras.layers.Dense(random.choice(neuron_num), activation='relu')(a)
  c = keras.layers.Dense(random.choice(neuron_num), activation='relu')(b)
  d = keras.layers.Dense(random.choice(neuron_num), activation='relu')(c)
  outs = keras.layers.Dense(random.choice(neuron_num), activation='sigmoid')(d)

  model = keras.Model(ins,outs)

  model.compile(loss=keras.losses.sparse_categorical_crossentropy,
              optimizer=random.choice(opt),
              metrics=[keras.metrics.sparse_categorical_accuracy])
  h = model.fit(X_train, y_train, epochs=random.choice(epoch_num), batch_size=random.choice(bsize) ,validation_data=(X_validation, y_validation))
  pd.DataFrame(h.history).plot(figsize=(12,8))
  plt.grid(True)
  plt.show()
  print(h.params)
  print(model.optimizer)
  print(b,c,d)
```
